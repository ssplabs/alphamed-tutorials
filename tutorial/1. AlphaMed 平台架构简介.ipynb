# 组件说明

## 模拟调试环境

模拟调试环境为 Notebook 代码调试提供底层支持，其本身独立运行，不依赖任何其它算法组件，并向其它算法组件提供协助工具，以帮助其它组件支持模拟调试功能。

## 合约消息组件

消息可以分为两类，一类用于和任务管理器交互，此类型消息格式固定，不支持自定义；另一类用于算法流程调度，此类型消息格式随算法而不同，支持自定义。

## 数据传输组件

数据传输组件提供工具接口，完成跨节点的数据传输，传输时仅依赖节点 ID 确定传输目的地，不使用 IP、域名。数据传输组件依赖*合约消息组件*沟通协调传输流程。

## AutoModel 组件

AutoML 代表 AlphaMed 平台自动建模机制，AutoModel 仅指此机制中的预训练模型概念。由于“预训练模型”是业界已有概念，且和 AlphaMed 平台中的预训练模型含义有差异，为避免混淆，使用 AutoModel 指代。
AutoModel 侧重于自动微调、自动部署、自动推理功能，不要求必须绑定联邦算法，因此其本身不依赖*数据传输组件*和*合约消息组件*。联邦 AutoModel 依赖上述两个组件是联邦算法的要求，本质上与 AutoModel 机制无关。

## 算法调度器组件

算法调度器用于管理算法流程，既可以是本地的，也可以是联邦的。所有算法调度器都必须继承共同的基础类 `Scheduler`，其提供了面向使用者的运行算法接口和面向任务管理器的启动计算进程接口，以及少量不依赖任何具体算法的通用工具。

### 内置横向联邦 FedAvg 算法家族

横向联邦 FedAvg 算法家族是一组算法调度器的实现，仅根据具体算法自动化控制算法流程（参数聚合），不依赖、不干涉任何具体模型。

### 内置异构联邦 HeteroNN 算法家族

异构联邦 HeteroNN 算法家族是一组算法调度器的实现，仅根据具体算法自动化控制算法流程（参数聚合），不依赖、不干涉任何具体模型。

### 交互式算法算子组件（敬请期待...）

交互式算法算子组件用于提供接口式的加密算法、多方安全计算算法等复杂算法调用，免去开发者自己实现的负担。对于仅需在本地执行的算法，比如传统的数据加密、加 DP 噪音等操作，均已有现成的成熟库函数可供直接使用，不是 AlphaMed 平台的主要目标。AlphaMed 的算子组件主要考虑需要多个参与方互相通信协作实现的算法，比如秘密分享、隐私求交、多方安全计算等。

### 自定义联邦算法实现和新增内置算法实现

二者本质相同，区别仅在于其实现代码是否内置于官方项目中。二者都必须继承 `Scheduler` 基础类，满足横向联邦或异构联邦任务管理器的管理要求，除此之外无其它要求。

## 联邦 AutoModel
联邦 AutoModel 本质上就是一个同时满足 AutoModel 机制和联邦算法调度器机制的实现，类似于多重继承。
